```python
def target_summary_with_num(dataframe, target, numerical_col):
    print(dataframe.groupby(target).agg({numerical_col: "mean"}), end="\n\n\n")
    dataframe[numerical_col].hist(by=dataframe[target])
    plt.xlabel(numerical_col)
    plt.ylabel("Frequency")
    plt.title("{} by {}".format(numerical_col, target))
    plt.show()
```

YÜKSEK KOLERASYON SAHİP DEĞERLERİN SİLİNMESİ

```python
cor_matrix =df.corr().abs()
```

```python
upper_triagle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
```

```python
drop_list = [col for col in upper_triagle_matrix.columns if any(upper_triagle_matrix[col] > 0.90)]
```

```python
df.drop(drop_list,axis=1,inplace=True)
```

```python
def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    corr = dataframe.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc={"figure.figsize":(15,15)})
        sns.heatmap(corr,annot=True,fmt=".2f",linewidths=0.5,square=True)
        plt.show()
    return drop_list
```

```python
high_correlated_cols(df,True,0.90)
```
